# Missing Values

```python
# to handle Categorical features which are missing

# Step 1) Find all the features with Null Values
categorical_with_na = [features for features in data.columns
                    if data[features].isnull().sum() > 1 
                    and data[features].dtypes=='O' # data type ~ O ~ Object ~ String
                   ]
categorical_with_na
```

```python
# Counting the Number of Null / missing Values in Categorical Features
data[categorical_with_na].isnull().sum()
```

```python
# Step 2) find percentage [dot.mean command] of Missing values in each feature
for feature in categorical_with_na:
    print(feature,'   :   ' ,
          np.round(data[feature].isnull().mean()*100,2),
          '%          of total Values are Missing')
    
# ",2" ~ rounding upto 2 decimals
# "dot.mean()" ~ to find the perecentage
```

```python
# Sep 3: to replace Nan / Missing Values with a New label

def replace_cat_feature(data,categorical_with_na):
    data2 = data.copy()
    data[categorical_with_na]=data[categorical_with_na].fillna('Missing') # replacing Nan / Missing Values with a New label
    return data

data = replace_cat_feature(data,categorical_with_na)

data[categorical_with_na].isnull().sum() # After Replacement, Counting the NUmber of Null / missing Values
```

# Missing Values in Numerical Features

```python
# To check the Numerical Features, which have Missing Values

# Step 1) Find all the features with Null Values
numerical_with_na = [features for features in data.columns
                    if data[features].isnull().sum() > 1 
                    and data[features].dtypes !='O' # data type ~ !='O' ~ Not Object ~ not String 
                                                                    # ~ Not categorical ~ Numerical
                   ]
numerical_with_na
```

```python
# Counting the Number of Null / missing Values in Categorical Features
data[numerical_with_na].isnull().sum()
```

```python
# Step 2) find percentage  of Missing values in each feature
for feature in numerical_with_na:
    print(feature,'   :   ' ,
          np.round(data[feature].isnull().mean()*100,2),
          '%          of total Values are Missing')
    
# ",2" ~ rounding upto 2 decimals
# "dot.mean()" ~ to find the perecentage
```

```python
# Sep 3: to replace Nan / Missing Values with a Median

for feature in numerical_with_na:
    median_value = data[feature].median() # defining "median_value"
    data[feature+'nan']=np.where(data[feature].isnull(),1,0) # defining new function
                                                            # to replace Nan values with 1 and others with 0
    data[feature].fillna(median_value,inplace=True) # replacing the New feature with Median

data[numerical_with_na].isnull().sum() # After Replacement, Counting the Number of Null / missing Values
```

# Handling Rare Categorical Feature
    - Remove Categorical Features which are present less than 1% of the observations

```python
data
```

```python
# Step 1) Find the List of all categorical features
categorical_features = [features for features in data.columns
                    if data[features].dtypes=='O' # data type ~ O ~ Object ~ String
                   ]
len(categorical_features)
```

```python
# Step 2) Calculate the Percentage of Number of values in each Categorical feature
for feature in categorical_features: # iterating for each one of the above features
    temp = data.groupby(feature)['SalePrice'].count()/len(data) # defining new variable 'temp'
                                                                # "groupby" ~ comparing (feature) with ['SalePrice']
    #temp = data.groupby(feature)['SalePrice'].mean()
    temp_df=temp[temp>0.01].index # Filtering features who have more than 1 % of values
    data[feature]=np.where(data[feature].isin(temp_df),data[feature],'Rare_Variable')
    # "isin(temp_df" condition ~ if data "is in [temp>0.01]" then take the value of feature

    
    # orelse just replace the value with new label 'Rare_Variable'
      
data[categorical_features].head(50)
# cltrl+F 
    # Rare_Variable
```

# How to convert Categorical features into Numerical features
    # for ML algorithms?

```python
for feature in categorical_features:
    labels_ordered=data.groupby([feature])['SalePrice'].mean().sort_values().index
    labels_ordered={k:i for i,k in enumerate(labels_ordered,0)}
    data[feature]=data[feature].map(labels_ordered)
```
